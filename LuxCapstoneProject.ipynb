```python
# Step 6: Load the data into python and assign a dataframe name
df = pd.read_csv('subscription_data.csv')
df.head()
```

```python
### DATA PREPARATION TASKS

# Step 7: Check the shape of the dataframe

df.shape
```

```python
# Step 8: Check the datatypes of the columns in the dataframe

df.info()
```

```python
# Step 9: Check for duplicated records in the dataframe
# No duplicate records were found in the dataframe
df.duplicated().sum()
```

```python
# Step 10: Check for missing values in the dataframe
# No missing values were found in the dataframe
df.isna().sum()
```

```python
# Step 11: Convert the Join_Date and Last_Login_Date columns from string to date datatypes

df['Join_Date'] = pd.to_datetime(df['Join_Date'], dayfirst=True)

df['Last_Login_Date'] = pd.to_datetime(df['Last_Login_Date'], dayfirst=True)
```

```python
# Step 12: Check the column datatypes to see if the 2 data columns have been converted to datetime format

df.info()
```

```python
# Step 13: Get today's date and assign it to a new column name(Analysis_Date) in the dataframe

today_date = pd.Timestamp.today().normalize()

today_date

df['Analysis_Date'] = today_date
```

```python
# Check to see if the new column name(Analysis_Date) has been created successfully
df.info()
```

```python
# Step 14: Create a new column name Last_Login_Duration showing days since last login

df['Last_Login_Duration'] = df['Analysis_Date'] - df['Last_Login_Date']

df['Last_Login_Duration'] = (df['Last_Login_Duration'].dt.total_seconds() / (24 * 3600)).astype(int)

df.head()
```

```python
# Step 15: Create a new column name Tenure showing duration in days since the joining date

df['Tenure'] = df['Analysis_Date'] - df['Join_Date']

df['Tenure'] = (df['Tenure'].dt.total_seconds() / (24 * 3600)).astype(int) # converts the column to an int datatype

df.rename(columns= {'Tenure' :'Tenure_in_Days'}, inplace=True)
```

```python
# Check to see if the new column name(Tenure) has been created successfully
df.info()
```

```python
# Step 16: Create a new column name (Total_Login_Days) showing duration in days between joining date and last login date

df['Total_Login_Days'] = df['Last_Login_Date'] - df['Join_Date']

df['Total_Login_Days'] = (df['Total_Login_Days'].dt.total_seconds() / (24 * 3600)).astype(int) # converts the column to a int datatype
```

```python
# Step 17: Create a new column(Monthly_Viewing_Days) by converting the Monthly_Viewing_Hours into days per month

df['Monthly_Viewing_Days'] = df['Monthly_Viewing_Hours'] / 24
df['Monthly_Viewing_Days'] = df['Monthly_Viewing_Days'].round(1)
df.head()
```

```python
#Step 18: Find the average number of viewing hours per day

df['Daily_Viewing_Hours'] = df['Monthly_Viewing_Hours'] / 30.44
df['Daily_Viewing_Hours'] = df['Daily_Viewing_Hours'].round(1)
df.head()
```

```python
# Step 19: Find the total number of months between the join date and last login date

df['Total_Login_Months'] = df['Total_Login_Days'] / 30.44
df['Total_Login_Months'] = df['Total_Login_Months'].round(1)
df.head()
```

```python
# Step 20: Find the total number of days that customers used the service

df['Total_Viewing_Days'] = df['Monthly_Viewing_Days'] * df['Total_Login_Months']

df['Total_Viewing_Days'] = df['Total_Viewing_Days'].round(1)

df.head()
```

```python
### EDA QUESTIONS

# Step 21: Chart a boxplot to check any outliers in the Monthly_Viewing_Hours column

df['Monthly_Viewing_Hours'].plot.box()
```

```python
# Step 22: Extract the outliers from the Monthly_Viewing_Hours and determine whether they should be removed

q1 = df['Monthly_Viewing_Hours'].quantile(0.25)
q2 = df['Monthly_Viewing_Hours'].quantile(0.50)
q3 = df['Monthly_Viewing_Hours'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['Monthly_Viewing_Hours'] < lower_bound)|(df['Monthly_Viewing_Hours'] > upper_bound)] 
outliers
```

```python
# Step 23: The outlier should not be removed as it is not vastly different from the rest of the values in the column

df.sort_values(by='Monthly_Viewing_Hours', ascending=False).head()
```

```python
# Step 24: Determine the Overall Churn Rate - 47.4%

Total_Initial_Customers = len(df)

Customers_Lost = df['Churn_Flag'].sum()

Churn_Rate = (Customers_Lost / Total_Initial_Customers) * 100

Churn_Rate

df.head()
```

```python
# Step 25: Analyze churn variation by subscription type
# The Premium plan had the highest average churn rate of 48.6% while the Basic plan had the lowest at 46.5%

churn_by_subscription_type = (df.groupby('Subscription_Type')['Churn_Flag'].mean() * 100).round(1)

churn_by_subscription_type
```

```python
# Step 26: Determine if there is a relationship between age and churn
# The young adults(18-35 years) had the highest average churn rate(49%) while the Seniors had the lowest (46%)

bins = [0, 17, 35, 50, 70] 

labels = ['Child', 'Youth', 'Adult', 'Senior']

df['Age_Category'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True) # Create a new column for age category

churn_by_age_category = (df.groupby('Age_Category')['Churn_Flag'].mean() * 100).round(0) #Find the average churn rate per age category

churn_by_age_category.plot.bar(title='Churn Rate by Age Category', color='red')
```

```python
bins = [0, 30, 60, 90] 

labels = ['Low', 'Moderate', 'High']

df['Monthly_Usage_Category'] = pd.cut(df['Monthly_Viewing_Hours'], bins=bins, labels=labels, include_lowest=True)
```

```python
# Step 27: Determine if there is a relationship between support tickets and churn rate
# Customers with low number of support tickets had the highest average churn rate (47.8%) while those with a high number of support tickets had the lowest(33.3%)

bins = [0, 2, 4, 6] 

labels = ['Low', 'Moderate', 'High']

df['Support_Ticket_Category'] = pd.cut(df['Support_Tickets'], bins=bins, labels=labels, include_lowest=True)

churn_by_support_tickets= (df.groupby('Support_Ticket_Category')['Churn_Flag'].mean() * 100).round(0)

churn_by_support_tickets

churn_by_support_tickets.plot.bar(x='Support_Ticket_Category', title= 'Churn Rate by Support Ticket Category', color='red')
```

```python
# Step 28: Determine if there is a difference in viewing behaviour between churned and retained customers
# The retained customers spent more viewing hours per month on average (40.26) than the churned customers (39.43)

Churned_Monthly_Viewing_Hours = (df.groupby('Churn_Flag')['Monthly_Viewing_Hours'].mean()).round(0)

Churned_Monthly_Viewing_Hours
```

```python
# Step 29: Determine whether there is a critical period after sign-up when churn risk is highest
# Churn risk was found to be highest in the 11th month after sign-up

Months_to_Churn = (df.groupby('Churn_Flag')['Total_Login_Months'].mean()).round(0)

Months_to_Churn.plot.bar(title= 'Average Duration to Churn')
```

```python
# Step 30: Determine whether payment method affects churn probability
# Paypal had the highest churn rate of 48% while credit card had the lowest at 46.43%

Payment_Method_Churn = (df.groupby('Payment_Method')['Churn_Flag'].mean() * 100).round(1)
Payment_Method_Churn
```

```python
# Step 31: Determine if use of discounts affected customer retention
# Customers who used the discount had less average churn rate (44.58%) than those that did not use the discount (50.30%)

churn_rate_by_discount = (df.groupby('Discount_Used')['Churn_Flag'].mean() * 100).round(1)

churn_rate_by_discount

churn_rate_by_discount.plot.bar(title= 'Churn Rate by Discount', color='red')
```

```python
# Step 32: Address the issue of the Total Login Days column showing some negative numbers

# find the records with the discrepancies between join date and last login date
df[df['Total_Login_Days'] < 0]

# show the discrepancies together with specific columns
odd_login_days = df[df['Total_Login_Days'] < 0][['Join_Date','Last_Login_Date', 'Total_Login_Days', 'Churn_Flag']]

odd_login_days

# Since the number of discrepancies found was small compared to the dataframe an assumption can be made that they were likely caused by human error during data entry

# Thus the solution should be to interchange the dates for only the records with the discrepancies

# define the condition for the discrepancies betwen the two dates
condition = df['Last_Login_Date'] < df['Join_Date'] 

# replace the join date with the last login date and vice versa for only those records with discrepancies
df.loc[condition, ['Join_Date', 'Last_Login_Date']] = (df.loc[condition, ['Last_Login_Date', 'Join_Date']].values)

# renew the column total login days to update the new changes to the discrepancies
df['Total_Login_Days'] = df['Last_Login_Date'] - df['Join_Date']

# convert the total login days column to an int datatype
df['Total_Login_Days'] = (df['Total_Login_Days'].dt.total_seconds() / (24 * 3600)).astype(int)
```

```python
# Step 33: Define churn risk factor conditions by identifying variables that have a clear relationship with churn rate

df[df['Churn_Flag'] == 1].shape

# Majority of the customers who churned (93%) had a low number of support tickets per customer (between 0 - 2 tickets)
df[(df['Support_Ticket_Category'] == 'Low') & (df['Churn_Flag'] == 1)].shape

# Majority of the customers who churned (92%) had either low or moderate average monthly viewing hours
df[(df['Monthly_Usage_Category'].isin(['Low', 'Moderate'])) & (df['Churn_Flag'] == 1)].shape

# Majority of the customers who churned (72%) belong to either the youth or senior age categories
df[(df['Age_Category'].isin(['Youth', 'Senior'])) & (df['Churn_Flag'] == 1)].shape

# Majority of the customers who churned (54%) had either free trial or premium subscriptions
#df[(df['Subscription_Type'].isin(['Free Trial', 'Premium'])) & (df['Churn_Flag'] == 1)].shape - not significant enough to be considered a churn risk factor
```

```python
# Step 34: Calculate the churn risk score and create a new column showing Churn Risk Levels

df['Churn_Risk_Score'] = 0

df.loc[df['Support_Ticket_Category'] == 'Low', 'Churn_Risk_Score'] += 1 # if the condition is true 1 is added to the risk score otherwise zero

df.loc[df['Monthly_Usage_Category'].isin(['Low', 'Moderate']), 'Churn_Risk_Score'] += 1

df.loc[df['Age_Category'].isin(['Youth', 'Senior']), 'Churn_Risk_Score'] += 1

# Categorize the churn risk scores into various levels

df['Churn_Risk_Level'] = pd.cut(df['Churn_Risk_Score'], bins=[0, 1, 2, 3], labels=['Low Risk', 'Moderate Risk', 'High Risk'], include_lowest=True)
```

```python
# Step 35: Determine the relationship between Average Churn Rate and Churn Risk Level

churn_risk_vs_churn_rate = (df.groupby('Churn_Risk_Level')['Churn_Flag'].mean() * 100).round(0).reset_index()

churn_risk_vs_churn_rate.rename(columns={'Churn_Flag': 'Churn_Rate(%)'}, inplace=True)

sns.barplot(x= 'Churn_Risk_Level', y= 'Churn_Rate(%)', data=churn_risk_vs_churn_rate)

plt.title('Churn Risk Level Versus Churn Rate')
```

```python
# Step 36: Determine the relationship between Average Monthly Viewing Hours and Churn Risk Level

churn_by_avg_hours = df.groupby('Churn_Risk_Level')['Monthly_Viewing_Hours'].mean().reset_index()
sns.barplot(x='Churn_Risk_Level', y='Monthly_Viewing_Hours', data=churn_by_avg_hours)
plt.title('Average Monthly Viewing Hours by Churn Risk Level')
```

```python
# Step 37: Determine the relationship between Average Daily Viewing Hours and Churn Risk Level

churn_risk_by_daily_hours = df.groupby('Churn_Risk_Level')['Daily_Viewing_Hours'].mean().round(1).reset_index()

sns.barplot(x='Churn_Risk_Level', y='Daily_Viewing_Hours', data=churn_risk_by_daily_hours)

plt.title('Average Daily Viewing Hours by Churn Risk Level')
```

```python
# Step 38: Determine the relationship between Average Churn Rate and the Discount Used

churn_rate_by_discount_used = (df.groupby('Discount_Used')['Churn_Flag'].mean() * 100).round(0).reset_index()

churn_rate_by_discount_used.rename(columns={'Churn_Flag': 'Churn_Rate(%)'}, inplace=True)

churn_rate_by_discount_used

sns.barplot(x='Discount_Used', y='Churn_Rate(%)', data=churn_rate_by_discount_used)

plt.title('Churn Rate by Discount Used')
```

```python
# Step 39: Determine the relationship between Average Churn Rate and the Subscription Type

churn_rate_by_subscription_type = (df.groupby('Subscription_Type')['Churn_Flag'].mean() * 100).round(0).reset_index()

churn_rate_by_subscription_type.rename(columns={'Churn_Flag': 'Churn_Rate(%)'}, inplace=True)

churn_rate_by_subscription_type

sns.barplot(x='Subscription_Type', y='Churn_Rate(%)', data=churn_rate_by_subscription_type, color='red')

plt.title('Churn Rate by Subscription Type')
```

```python
# Step 40: Determine the relationship between Churn Risk Level and the Subscription Type

churn_risk_per_subscription_type = (df.groupby('Subscription_Type')['Churn_Flag'].mean() * 100).round(0).reset_index()

churn_risk_per_subscription_type.rename(columns={'Churn_Flag': 'Churn_Rate(%)'}, inplace=True)

churn_risk_per_subscription_type = churn_risk_per_subscription_type.merge(
    df[['Subscription_Type', 'Churn_Risk_Level']],
    on='Subscription_Type',
    how='left'
)

sns.barplot(data=churn_risk_per_subscription_type, x='Subscription_Type', y='Churn_Rate(%)', hue='Churn_Risk_Level')

plt.title('Churn Risk per Subscription Type')
plt.ylabel('Churn Rate (%)')  # Show percentage label on Y-axis
plt.xlabel('Subscription Type')
plt.show()

```

```python
# Step 41: Determine the relationship between Churn Risk Level and the Auto Renewal Status

churn_rate_by_auto_renewal = (df.groupby('Auto_Renew_Enabled')['Churn_Flag'].mean() * 100).round(0).reset_index()

churn_rate_by_auto_renewal.rename(columns={'Churn_Flag': 'Churn_Rate(%)'}, inplace=True)

sns.barplot(x='Auto_Renew_Enabled', y='Churn_Rate(%)', data=churn_rate_by_auto_renewal)

plt.title('Churn Rate by Auto Renewal Status')
```
